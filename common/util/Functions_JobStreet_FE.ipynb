# Function#2 - Preliminary cleaning of data

```python
# BEFORE
df[['company_address','company_employee_num','company_benefit','company_dress_code','company_language','company_work_hours']]
```

```python
def prelim_clean(DataFrame):
    import re
    
    df = DataFrame
    for x in df.index:
        
#         #JS COL
#         if 'https' in df.loc[x, 'company_js_url']:
#             df.loc[x, 'company_js_url'] = re.sub(r'(https.*companies)', '', df.loc[x, 'company_js_url'])

        #LOCATION COL
        if 'Location' in df.loc[x, 'company_address']:
            df.loc[x, 'company_address'] = re.sub(r'Location', '', df.loc[x, 'company_address'])

        #WEBSITE COL
#         if 'Website' in df.loc[x, 'company_website']:
#             df.loc[x, 'company_website'] = re.sub(r'Website', '', df.loc[x, 'company_website'])

        #COMPANY SIZE COL
        if 'Company Size' in df.loc[x, 'company_employee_num']:
            df.loc[x, 'company_employee_num'] = re.sub(r'Company Size', '', df.loc[x, 'company_employee_num'])
        if 'Employees' in df.loc[x, 'company_employee_num']:
            df.loc[x, 'company_employee_num'] = re.sub(r'Employees', '', df.loc[x, 'company_employee_num'])
        if 'More than' in df.loc[x, 'company_employee_num']:
            df.loc[x, 'company_employee_num'] = re.sub(r'More than', '>', df.loc[x, 'company_employee_num'])

        #BENEFITS COL
        if 'Benefits' in df.loc[x, 'company_benefit']:
            df.loc[x, 'company_benefit'] = re.sub(r'Benefits', '', df.loc[x, 'company_benefit'])

        #DRESS CODE COL
        if 'Dress code' in df.loc[x, 'company_dress_code']:
            df.loc[x, 'company_dress_code'] = re.sub(r'Dress code', '', df.loc[x, 'company_dress_code'])

        #LANGUAGE COL
        if 'Spoken language' in df.loc[x, 'company_language']:
            df.loc[x, 'company_language'] = re.sub(r'Spoken language', '', df.loc[x, 'company_language'])

        #WORK HOURS COL
        if 'Work hours' in df.loc[x, 'company_work_hours']:
            df.loc[x, 'company_work_hours'] = re.sub(r'Work hours', '', df.loc[x, 'company_work_hours'])
            
    return df
```

```python
# AFTER
df = prelim_clean(df)
df[['company_address','company_employee_num','company_benefit','company_dress_code','company_language','company_work_hours']]
```

# Function#3 - Classifying Address by States

```python
# BEFORE
df[['company_address']]
```

```python
def classify_state(DataFrame):
    df = DataFrame
    
    if "company_state" not in df.columns:
        df.insert(10, "company_state", "")  

    KL = ['kuala lumpur', 'kl', 'bukit jalil', 'wilayah persekutuan', 'q sentral']
    KD = ['kedah']
    SL = ['selangor', 'klang', 'puchong', 'petaling jaya', 'ampang', 'subang', 'shah alam',
          'damansara', 'petaling', 'cheras', 'kelana jaya','sungai buloh', 'tropicana avenue', 
          'pj']
    PG = ['penang', 'pinang', 'georgetown']
    JH = ['johor', 'pasir gudang']
    PR = ['perak']
    PH = ['pahang']
    NS = ['negeri sembilan', 'seremban', 'port dickson']
    SR = ['sarawak', 'kuching']
    ML = ['malacca', 'melaka', 'alor gajah']
    SB = ['sabah']
    KT = ['kelantan']
    SG = ['singapore']

    for x in df.index:

        #LOWER-CASING
        df.loc[x, 'company_address'] = df.loc[x, 'company_address'].lower()

        #KUALA LUMPUR
        for y in range(len(KL)):
            if KL[y] in df.loc[x, 'company_address']:
                #df.loc[x, 'company_address'] = 'None'
                df.loc[x, 'company_state'] = "Kuala Lumpur"

        #SELANGOR
        for y in range(len(SL)):     
            if SL[y] in df.loc[x, 'company_address']:
                #df.loc[x, 'company_address'] = 'None'
                df.loc[x, 'company_state'] = "Selangor"

        #KEDAH    
        for y in range(len(KD)):    
            if KD[y] in df.loc[x, 'company_address']:
                #df.loc[x, 'company_address'] = 'None'
                df.loc[x, 'company_state'] = "Kedah"

        #JOHOR
        for y in range(len(JH)):    
            if JH[y] in df.loc[x, 'company_address']:
                #df.loc[x, 'company_address'] = 'None'
                df.loc[x, 'company_state'] = "Johor"

        #PENANG
        for y in range(len(PG)):
            if PG[y] in df.loc[x, 'company_address']:
                #df.loc[x, 'company_address'] = 'None'
                df.loc[x, 'company_state'] = "Penang"

        #SABAH
        for y in range(len(SB)):
            if SB[y] in df.loc[x, 'company_address']:
                #df.loc[x, 'company_address'] = 'None'
                df.loc[x, 'company_state'] = "Sabah"

        #PAHANG
        for y in range(len(PH)):    
            if PH[y] in df.loc[x, 'company_address']:
                #df.loc[x, 'company_address'] = 'None'
                df.loc[x, 'company_state'] = "Pahang"

        #PERAK
        for y in range(len(PR)):          
            if PR[y] in df.loc[x, 'company_address']:
                #df.loc[x, 'company_address'] = 'None'
                df.loc[x, 'company_state'] = "Perak"

        #SARAWAK
        for y in range(len(SR)):          
            if SR[y] in df.loc[x, 'company_address']:
                #df.loc[x, 'company_address'] = 'None'
                df.loc[x, 'company_state'] = "Sarawak"

        #MELAKA
        for y in range(len(ML)):          
            if ML[y] in df.loc[x, 'company_address']:
                #df.loc[x, 'company_address'] = 'None'
                df.loc[x, 'company_state'] = "Melaka"

        #NEGERI SEMBILAN
        for y in range(len(NS)):          
            if NS[y] in df.loc[x, 'company_address']:
                #df.loc[x, 'company_address'] = 'None'
                df.loc[x, 'company_state'] = "Negeri Sembilan"

        #KELANTAN
        for y in range(len(KT)):          
            if KT[y] in df.loc[x, 'company_address']:
                #df.loc[x, 'company_address'] = 'None'
                df.loc[x, 'company_state'] = "Negeri Sembilan"

        #SINGAPORE
        for y in range(len(SG)):          
            if SG[y] in df.loc[x, 'company_address']:
                #df.loc[x, 'company_address'] = 'None'
                df.loc[x, 'company_state'] = "Singapore"
    return df
```

```python
# AFTER
df = classify_state(df)
df[['company_address','company_state']]
```
